```python
from datasets import load_dataset
```

```python
ds = load_dataset(
    "alkzar90/NIH-Chest-X-ray-dataset", 
    'image-classification',
    num_proc = 12
)
```

# Subset 10k

```python
subset_size = 10_000
ds = ds["train"].shuffle(seed=7).select(range(subset_size))
ds = ds.train_test_split()
```

```python
def preprocess_single(item, resizeTo=(300, 300)):
    # not all images are RGB     
    item["image"] = item["image"].resize(resizeTo).convert('RGB')
    return item

ds = ds.map(preprocess_single, batched=False, num_proc=12)
```

```python
ds.push_to_hub("g-ronimo/" + f"NIH-Chest-X-ray-dataset_{round(subset_size/1000)}k")
```

# Full Resized

```python
def preprocess_single(item, resizeTo=(300, 300)):
    # not all images are RGB     
    item["image"] = item["image"].resize(resizeTo).convert('RGB')
    return item

ds = ds.map(preprocess_single, batched=False, num_proc=6)
```

```python
ds.push_to_hub("g-ronimo/NIH-Chest-X-ray-dataset_resized300px")
```
